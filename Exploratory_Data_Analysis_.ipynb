```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
sales_data = pd.read_csv('actuals.csv')

sales_data.head()
```

```python
sales_data.isnull().sum()
```

```python
sales_data.shape
```

DATA VISUALIZATION

```python
sales_data.hist(figsize=(8,9))
```

```python
sales_data.plot()
```

```python
sns.countplot(sales_data['Product'])
sales_data['Product'].value_counts()
```

```python
sns.scatterplot(sales_data['Sales'], sales_data['Category'])
```

```python
sns.boxplot(x=sales_data['Sales'], y=sales_data['Boxes'])
```

```python
sns.heatmap(np.round(sales_data.corr(),2), annot=True)
```

```python
plt.figure(figsize=(12, 6))
plt.title('Total Monthly transaction by teams')
sns.countplot(sales_data['Product'], hue = sales_data.Team)
```
